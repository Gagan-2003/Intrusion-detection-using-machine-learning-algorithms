```python
from sklearn import datasets
import numpy as nm
import matplotlib.pyplot as mtp
import pandas as pd
```

```python
df['intrusion']=df['effect'].apply(lambda x: 1 if x=='anomaly' else 0)
df
```

```python
inputs = df.drop('intrusion',axis='columns')
target =df['intrusion']
```

```python
from sklearn.preprocessing import LabelEncoder
le_protocol_type = LabelEncoder()
le_service = LabelEncoder()
le_flag = LabelEncoder()
```

```python
inputs['protocol_type_n'] = le_protocol_type.fit_transform(inputs['protocol_type'])
inputs['service_n'] = le_protocol_type.fit_transform(inputs['service'])
inputs['flag_n'] = le_protocol_type.fit_transform(inputs['flag'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['effect','protocol_type','duration','service','flag','src_bytes'],axis='columns')
```

```python
inputs_n
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict(inputs_n[0:20])
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(inputs_n)
cm = confusion_matrix(target, y_pred)
cm
```
