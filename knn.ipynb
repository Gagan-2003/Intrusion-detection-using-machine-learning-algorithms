```python
from sklearn import datasets
import numpy as nm
import matplotlib.pyplot as mtp
import pandas as pd
```

```python
df['intrusion']=df['effect'].apply(lambda x: 1 if x=='anomaly' else 0)
df
```

```python
inputs = df.drop('intrusion',axis='columns')
target =df['intrusion']
```

```python
from sklearn.preprocessing import LabelEncoder
le_protocol_type = LabelEncoder()
le_service = LabelEncoder()
le_flag = LabelEncoder()
```

```python
inputs['protocol_type_n'] = le_protocol_type.fit_transform(inputs['protocol_type'])
inputs['service_n'] = le_protocol_type.fit_transform(inputs['service'])
inputs['flag_n'] = le_protocol_type.fit_transform(inputs['flag'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['effect','protocol_type','service','flag','hot','num_failed_login','logged_in','land','wrong_fragment','num_compromised','root_shell','su_attempted','num_root','num_file_creations','num_shells','num_access_files','num_outbound_cmds','is_host_login','is_guest_login','srv_count','serror_rate','srv_serror_rate','rerror_rate','srv_rerror_rate','same_srv_rate','diff_srv_rate','srv_diff_host_rate','dst_host_count','dst_host_srv_count','dst_host_same_srv_rate','dst_host_diff_srv_rate','dst_host_same_src_port_rate','dst_host_srv_diff_host_rate','dst_host_serror_rate','dst_host_srv_serror_rate','dst_host_rerror_rate','dst_host_srv_rerror_rate'],axis='columns')
```

```python
inputs_n
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```
